```python
a= dataset[dataset['Provincia']=='Granada']



conteo= len(a)

suma= sum(a['Poblacion'])

suma

#Promedio

promedio= suma/conteo

promedio


median_value = np.percentile(a['Poblacion'], 50)

max= max(a['Poblacion'])
min= min(a['Poblacion'])

st_dev = np.std(a['Poblacion'])

print(promedio)
print(median_value)
print(st_dev)
print(max)
print(min)



```
